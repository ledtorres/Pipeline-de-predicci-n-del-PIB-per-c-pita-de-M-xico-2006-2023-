# Pipeline predictivo del PIB per cápita estatal en México: integrando educación, inversión, remesas y violencia (2006-2023)
# Merge de Datos

Importación de líbrerias y datos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
pibe = pd.read_csv("pibe_per_capita_estados.csv")
informalidad = pd.read_csv("informalidad_estados.csv")
escolaridad = pd.read_csv("escolaridad_estados.csv")
ied = pd.read_csv("ied_estados.csv")
remesas = pd.read_csv("remesas_estados.csv")
violencia = pd.read_csv("violencia_estados.csv")
sectorial = pd.read_csv("composicion_sectorial.csv")
```

Chequeo de datos

```python
dfs = {
    'pibe': pibe,
    'informalidad': informalidad,
    'escolaridad': escolaridad,
    'ied': ied,
    'remesas': remesas,
    'violencia': violencia,
    'sectorial': sectorial}

for nombre, df in dfs.items():
    print(f"{nombre}: {df.shape} | años: {df['año'].min()}-{df['año'].max()} | estados: {df['estado'].nunique()} | nulls: {df.isnull().sum().sum()}")
```

Combinación de los documentos en un solo dataset

```python
AÑO_INICIO = 2006
AÑO_FIN = 2023

pibe = pibe[pibe['año'].between(AÑO_INICIO, AÑO_FIN)]
sectorial = sectorial[sectorial['año'].between(AÑO_INICIO, AÑO_FIN)]
informalidad = informalidad[informalidad['año'].between(AÑO_INICIO, AÑO_FIN)]
escolaridad = escolaridad[escolaridad['año'].between(AÑO_INICIO, AÑO_FIN)]
ied = ied[ied['año'].between(AÑO_INICIO, AÑO_FIN)]
remesas = remesas[remesas['año'].between(AÑO_INICIO, AÑO_FIN)]
violencia = violencia[violencia['año'].between(AÑO_INICIO, AÑO_FIN)]
```

```python
rename_map = {
    'Veracruz': 'Veracruz de Ignacio de la Llave',
    'Estado de México': 'México',
    'Coahuila': 'Coahuila de Zaragoza',
    'Michoacán': 'Michoacán de Ocampo',
    'Distrito Federal': 'Ciudad de México'
}

remesas.loc[:, 'estado'] = remesas['estado'].replace(rename_map)
ied.loc[:, 'estado'] = ied['estado'].replace(rename_map)
informalidad.loc[:, 'estado'] = informalidad['estado'].replace(rename_map)
```

```python
df = pibe.copy()
df = df.merge(sectorial, on=['estado', 'año'], how='left')
df = df.rename(columns={'pibe_per_capita': 'pib_per_capita'})
df = df.merge(informalidad, on=['estado', 'año'], how='left')
df = df.merge(escolaridad, on=['estado', 'año'], how='left')
df = df.merge(ied, on=['estado', 'año'], how='left')
df = df.merge(remesas, on=['estado', 'año'], how='left')
df = df.merge(violencia, on=['estado', 'año'], how='left')
```

```python
df.to_csv('merged_data.csv', index=False)
print('Merged data saved to merged_data.csv')
```
